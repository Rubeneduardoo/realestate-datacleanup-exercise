#### Exercise 09. Are the average prices of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about them.

```python
# TODO
#Average Price Estimation for Valdemorillo and Galapagar
a = ds.loc[ds['address'] == 'Valdemorillo']['price'].mean()
b = ds.loc[ds['address'] == 'Galapagar']['price'].mean()

print(f'El precio medio para Valdemorillo es',a, 'mientras que para Galapagar es',b)

c = pd.Series([a,b])


#Conclusion
print('A pesar que, los precios promedios de Valdemorillo y Galapagar no son exactamente iguales, la desviación estandar entre ambos es de USD', c.std(), 'significando apenas una diferencia relativa de', ((c.std()/c.mean())*100),'%','haciendo que, a la percepción pública ambos precios se consideren aproximadamente igual')
```

#### Exercise 10. Are the average prices per square meter (price/m2) of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about it.

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
# TODO

#Creamos la nueva columna price/m2
ds['pps'] = (ds['price'] / ds['surface'])

#Calculamos el promedio para cada localidad
ppsV = ds.loc[ds['address'] == "Valdemorillo"]['pps'].mean()
ppsG = ds.loc[ds['address'] == "Galapagar"]['pps'].mean()

d= pd.Series([ppsV,ppsG])

print(f'Los precios por metro cuadrado para Valdemorillo son inferiores en comparacion a Galapagar. Para el primero, el precio por m2 es de:', ppsV, 'mientras que, para el segundo es de', ppsG,'haciendo a Valdemorillo una opcion mas economica por m2 que Galapagar')
print(f'Asimismo, la diferencia relativa de USD por m2 entre ambos es de un', (d.std()/d.mean())*100,'%')
```
